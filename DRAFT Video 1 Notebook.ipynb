# <font color='#183766'>Introduction to Jupyter Notebooks and JupyterLab</font>

In this series, we will use **Jupyter Notebooks** to run our Python code. Why use these notebooks?
- *With Jupyter Notebooks, you can write code **side-by-side with normal text**, allowing you to provide headings, descriptors, and notes to accompany your code for easier readability.*
- *Notebooks also **splits lines of code into individual cells**, which is great for running only certain code and breaking code up into reasonable or logical sections.*

These features make Jupyter Notebooks extremely useful and favorable for data scientists, and learning how to use these Notebooks is a valuable skill.

In fact, *this* document is a Jupyter Notebook!

Jupyter Notebooks use the file extension `.ipynb`. We can use **JupyterLab** to open, create, and modify Jupyter Notebooks. 

#### **How to Open JupyterLab**

1) First, **open the Anaconda Navigator** application that you installed in the previous section. 

2) Find the tile with the title `Jupyter Lab`, and press the `Launch` button. This launches JupyterLab in your browser!

JupyterLab is browser-based, however locally run. This means that bookmarking JupyterLab in your browser will not necessarily allow you to open JupyterLab later unless Anaconda is running at the same time.

# <font color='#183766'>Creating Cells</font>

The core building blocks of Jupyter Notebooks are **cells**. Cells can contain code and text based on the type of the cell, which can be edited on the toolbar at the top! It is likely that when you create a Jupyter Notebook, the first cell will likely be the `Code` cell type. There are three types of code types.

#### **Code Cells**

Code cells are cells that contain Python code! An example code cell with a single line of Python code is shown below:

```python
print("Hello, world!")
```

Jupyter also provides basic **syntax highlighting**, which is color-coding for your code! You will learn more about what each color refers to later.

This particular cell of code also produced **output**, or information that is presented back to the user. You can see output right below the cell! 

In this case, `Hello, world!` was the output for the cell.

There is also a number in brackets `[  ]` that are next to each cell. This provides the order in which the cells were run. You can run a cell by either pressing on the "run" icon on the toolbar, or pressing `Shift+Enter` on your keyboard.

You can also reset all cells by pressing on the reset button in the toolbar. This will not delete you wrote, however removes all outputs and allows you to rerun your code from the beginning.
