# <font color='#183766'>Introduction to Variables</font>

No matter which website, app, or program you use, that program must keep track of some type of data. For a game, this might be a high score, or for a notes app, the notes that you write! In addition, without being able to keep track of data, there would be no data science.

In Python, we can store data in **variables**. Variables allow us to assign a *value* to a *name* by which we can refer to the value stored later! We can also change this data if necessary later on. Variables are a fundamental building block of any programming language.

Variables are like a **container**, or a box. In the container, you put something (some value) that you might want to access or modify later. Outside the container, you might put some sort of label (name) so that when you are looking through your containers, you will be able to find the right one! 

In addition, just like with containers in real life, the **type** of container you use to store certain things might matter. For example, for storing a liquid, using a container with holes would make no sense! However, if transporting an animal such as a hamster, holes are necessary. We will touch on this concept in a little bit, but hold on to this thought!

First, this is how to create a variable in Python!

#### **Creating a Variable**

In order to create a variable, you must follow this basic framework:

```python
name = value
```

First, you must specify the `name` of the variable! You will then be able to refer back to the variable using this name later.

Following the `name` comes the assignment operator (`=`). The assignment operator assigns the `value` on the right to the name on the left! 

To the right of the assignment operator is the `value` that you actually want to store inside your variable.

#### **Choosing Variable Names**

When naming your variable, there are certain rules. For example:
- Variable names **cannot include a space in them.** If you want to name variables with a space, it is convention in Python to use an underscore (_). For example, an invalid variable name would be `my variable`, however `my_variable` is valid!
- Variable names **cannot include special characters and symbols**, as these are often reserved for other things in Python! Refrain from using these in your variable names.
- It is also convention for variable names in Python to be **lowercase**. This is because names starting with a capital letter are often reserved for classes, which is a topic we will not need to cover for now.

#### **Displaying Variable Values**

There are two main ways to display the value of variables in Python. First, if you put the name of a variable as the *last line of code in your cell*, the value of the variable is printed out! For example:

```python
x = 8
x
```

However, there are many times in which you will want to display the value of a variable in the middle of your code. You can use the `print()` function anywhere in your code to do this. To use the `print()` function, simply put the variable name between the parenthesis. For example:

```python
x = 8
print(x)
y = 4
print(y)
```

This code outputs `8` for the `print(x)` line, and `4` for the `print(y)` line.

# <font color='#183766'>Basic Data Types</font>

What type of values can you actually store in variables? There are many, however there are the four most basic ones that will be covered in this lesson. Remember the analogy about the type of container from above? This applies directly to data types! You must make sure that the data type of your variables will best hold the type of data you want to store.

#### **Integers (`int`)**

Integers are variables whose values are **numbers without any decimal places!** This includes all integers, positive and negative. The following is an example of an integer variable:

```python
my_int = 2
my_int
```

```python
my_second_int = -64
my_second_int
```

```python
zero = 0
print(zero)
```

#### **Floats (`float`)**

Floats are variables whose values are **numbers with decimal places!** This differs from an integer. For example:

```python
my_float = 2.3
my_float
```

```python
approx_p = 3.141592
print(approx_p)
```

*Challenge: What type do you think the variable below is?*

```python
mystery_var = 8.0
```

If you guessed a *double*, you are correct! Despite `8.0` being equivalent to `8` as an integer, just having `.0` as a decimal place will make Python interpret the number as a float. **We can verify this using the `type()` function**, which prints out the data type of the variable we put between the parenthesis:

```python
type(mystery_var)
```

#### **Booleans (`bool`)**

Booleans are a data type that has only two possible values -- `True`, or `False`! Nothing else. Booleans are a binary choice, meaning they represent only one of two states. For example:

```python
my_bool = True
my_second_bool = False
print(my_bool)
print(my_second_bool)
```

#### **Strings (`str`)**

You may have been wondering - we covered numbers, what about text? Well, strings are the data type that hold text!

It is extremely important to note that with strings, all text **must** be surrounded by quotation marks (`"`)! This allows Python to know the difference between a string and different variable names. For example, this is a correct definition of a string:

```python
my_name = "Ajay"
print(my_name)
```

It is common convention in Python to use single-quotes rather than double quotes to surround parenthesis, however both are acceptable! You will see it both ways:

```python
cads = 'Carolina Analytics and Data Science'
print(cads)
```
